 ## Can You Survive The Titanic?
 ### 1. Clean the dataset provided in `titanic_train_data.csv`
 Since this is a small dataset, I have generated an performed manual observation of the dataset and noticed columns 'Cabin' and 'Age' consist significant amount of 'Nan' in their respective data. This would negatively affect the quality of dataset and subsequent results of data analysis.
 

```python
import numpy as np 
import pandas as pd
from numpy import genfromtxt
from IPython.display import display
import array
import math
import matplotlib.pyplot as plt


in_file = 'titanic_train_data.csv'
full_data = pd.read_csv(in_file)
cabin_data = full_data['Cabin']
cabin_list = cabin_data.tolist()

x = []
i=0
for i in range(len(cabin_list)):
	if type(cabin_list[i]) == float:
		x.append(cabin_list[i])
	i=i+1
print(len(cabin_list))
print(len(x))
display(full_data.head())

```

After running the code above, I have observed that out of 891 data in the column 'Cabin', 687 of them were 'NaN'. That is a significant 77.1% of the entire data in column 'Cabin'. Therefore this would negatively affect quality of data. 
Hence the next step would be to remove column labeled 'Cabin'.

```python
new_full_data = full_data.drop('Cabin',1)
display(new_full_data.head())
```

Next column with significant amount of 'NaN' would be 'Age' column

```python
age_data = full_data['Age']
age_list = age_data.tolist()
print(len(age_list))
y= []
j=0
for j in range(len(age_list)):
    if np.isnan(age_list[j]):
        y.append(age_list[j])
    j=j+1
print(len(y))
age_array = np.asarray(age_list)
age_array[np.isnan(age_array)] = 0
print(len(age_array))
print(np.sum(age_array))
mean_age = (np.sum(age_array)/(len(age_array)-len(y)))
display(mean_age)
```

A simple code to run analysis the column 'Age' shows that there are a total of 177 'NaN' out of a total of 891 data in the column. That is an approximate of 19.9% of the entire data in the column. Therefore I would obtain the mean of age column to replace the 'NaN' in the 'Age column.

```python

npfullinds = np.where(np.isnan(new_full_data['Age']))
# # full_dataarraynp = np.asarray(new_full_data)
full_dataarray = pd.DataFrame(new_full_data)
y = full_dataarray.fillna(mean_age)
display(y)
```

### 2. Describe the new dataset.
The cleaned dataset now contains less bad data which would affect the quality of results of analysis. This was achieved by removing the 'NaN' majority column of Cabin column and by replacing 'NaN' age column with the mean age of the rest of column in order to make the dataset more meaningful for analysis.

```python
final_age_distribution = y.Age.value_counts()
display(final_age_distribution)
```

## Getting The Guts From A Fairy Tale

### 1. Clean `caroll-alice.txt` as clean as you can.

I first fetch data from caroll-alice.txt and split text into words and remove all tokens that are not alphabetic 
I then make every word lowercased and filter out stopwords provided.

```python


filename = 'caroll-alice.txt'
file = open(filename, 'rt')
text = file.read()
file.close()
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
words = [word for word in tokens if word.isalpha()]
final = [w.lower() for w in words]
provided_stop_words_file = 'stopwords.txt'
provided_stop = open(provided_stop_words_file, 'rt')
provided_stop_words = provided_stop.read()
provided_stop.close()
nostopwords = [w for w in final if not w in provided_stop_words]
print(nostopwords)
```
